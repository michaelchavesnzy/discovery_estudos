```python
import requests
import pandas as pd
```

## Get in requests

```python
#API do IBGE sobre localidades (Pais, estado, cidades)
url_base = "https://servicodados.ibge.gov.br/api/v1/localidades"
```

```python
endpoint_paises = url_base+"/paises"

response = requests.get(endpoint_paises)
```

```python
response.status_code
```

```python
response.json()
```

```python
df_paises = pd.json_normalize(response.json(), max_level = 0)
df_paises
```

```python
endpoint_estados = url_base+"/estados"

response = requests.get(endpoint_estados)
```

```python
pd.json_normalize(response.json(), max_level=0)
```

```python
#https://servicodados.ibge.gov.br/api/v1/localidades/estados/{UF}/municipios

endpoint_municipios_ceara = url_base+"/estados/23/municipios"

response = requests.get(endpoint_municipios_ceara)
```

```python
df_municipios_ceara = pd.json_normalize(response.json(), max_level=0)
df_municipios_ceara
```

```python
df_municipios_ceara[df_municipios_ceara['nome'] == 'Fortaleza']
```
